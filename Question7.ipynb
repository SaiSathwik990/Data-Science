```python
#import require python classes and packages
import pandas as pd #pandas to read and explore dataset
import matplotlib.pyplot as plt #use to visualize dataset vallues
import numpy as np
```

```python
#loading and displaying IRIS dataset with default single indexing
dataset = pd.read_csv("iris.csv")
dataset
```

```python
#hierarchical indexing
dataset.set_index(['SepalLengthCm', 'SepalWidthCm'], inplace=True)
dataset
```

```python
#handling missing values
dataset = pd.read_csv("iris.csv")
dataset.fillna(0, inplace = True)
dataset.isnull().sum()
```

```python
#arithmetic and boolean operations
#describing dataset with details like count, mean, standard deviation of each dataset attributes
dataset.describe()
```

```python
#pandas boolean operations
#read all rows whose sepal length > 6.8 and < 7
data = dataset[(dataset['SepalLengthCm'] < 7) & (dataset['SepalLengthCm'] > 6.8)]
data
```

```python
#pandas merging and aggregation operations
dataset.aggregate(['sum', 'min'])
```

```python
#Plotting single column graph 
#visualizing different class labels found in dataset
labels = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
names, count = np.unique(dataset['Species'].ravel(), return_counts = True)
height = count
bars = labels
y_pos = np.arange(len(bars))
plt.figure(figsize = (6, 3)) 
plt.bar(y_pos, height)
plt.xticks(y_pos, bars)
plt.xlabel("Dataset Class Label Graph")
plt.ylabel("Count")
plt.show()
```

```python
#plotting entire table graph
dataset.hist(figsize=(10, 8))
plt.title("Representation of Dataset Attributes")
plt.show()
```
